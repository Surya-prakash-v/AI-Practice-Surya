```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

## Edge Detector

```
edges = cv2.Canny(image,100,200)
cv2_imshow(edges)

edgekernal = np.float32([[-1,-1,-1],
                         [-1,8,-1],
                         [-1,-1,-1]])
edge2 = cv2.filter2D(image,-1,edgekernal)
cv2_imshow(edge2)
```

## Vertical Edge Detector

```
kernel = np.float32([[-1,0,1],
                     [-1,0,1],
                     [-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

## Horizantal Edge Detector

```
kernel = np.float32([[-1,-1,-1],
                     [0,0,0],
                     [1,1,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

## 45 Degree Angle Detector

```
kernel45 = np.float32([[-1,0,1],
                       [0,0,0],
                       [1,0,-1]])

kernel45reverse = np.float32([[1,0,-1],
                              [0,0,0],
                              [-1,0,1]])

dst45 = cv2.filter2D(image,-1,kernel45)
cv2_imshow(dst45)

dst45reverse = cv2.filter2D(image,-1,kernel45reverse)
cv2_imshow(dst45reverse)
```

## Blur Kernel

```
blurimg = cv2.blur(image,(10,10))
cv2_imshow(blurimg)

blurkernal = np.float32([[1,1,1],
                         [1,1,1],
                         [1,1,1]])/9
blurimg2 = cv2.filter2D(image,-1,blurkernal)
cv2_imshow(blurimg2)
```

## Sharpen Kernel

```
sharpenkernel = np.float32([[-1,-1,-1],
                            [-1, 9,-1],
                            [-1,-1,-1]])

sharpimg = cv2.filter2D(image,-1,sharpenkernel)
cv2_imshow(sharpimg)
```

## Identity function

```
identitykernel = np.float32([[0,0,0],
                             [0,1,0],
                             [0,0,0]])

identity = cv2.filter2D(image,-1,identitykernel)
cv2_imshow(identity)
```
